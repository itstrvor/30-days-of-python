```python
#Calculate slope and Euclidean distance
import math
x1,y1=2,2
x2,y2=6,10
#Find slope
def slope(x1,x2,y1,y2):
    m=(y2-y1)/(x2-x1)
    return m
#Euclidean distance
def eucl_distance(x1,x2,y1,y2):
    distance=math.sqrt((x2-x1)**2+(y2-y1)**2)
    return distance

x1,y1=2,2
x2,y2=6,10

print("slope(m)=",slope(x1,x2,y1,y2))
print("Euclidean distance=",eucl_distance(x1,x2,y1,y2))

```

```python
#  Most Spoken Languages

import json
from collections import Counter

def most_spoken_languages(filename, n):
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            countries = json.load(f)
    except FileNotFoundError:
        return f"Error: File '{filename}' not found."
    except json.JSONDecodeError:
        return f"Error: File '{filename}' contains invalid JSON."

    lang_counter = Counter()
    for c in countries:
        for lang in c.get("languages", []):
            lang_counter[lang] += 1

    return lang_counter.most_common(n)

# Run example
print(most_spoken_languages('./data/countries.json', 5))
```

```python
# Most Populated Countries

def most_populated_countries(filename, n):
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            countries = json.load(f)
    except FileNotFoundError:
        return f"Error: File '{filename}' not found."
    except json.JSONDecodeError:
        return f"Error: File '{filename}' contains invalid JSON."

    filtered = [
        {
            "country": c.get("name", "Unknown"),
            "population": c.get("population", 0)
        }
        for c in countries
    ]

    sorted_list = sorted(filtered, key=lambda x: x['population'], reverse=True)
    return sorted_list[:n]

# Run example
for country in most_populated_countries('./data/countries.json', 5):
    print(country)
```

```python
# Cat API — weight and lifespan stats
cats_api = 'https://api.thecatapi.com/v1/breeds'

resp = requests.get(cats_api)
cats = resp.json()

weights = []
lifespans = []

for cat in cats:
    # weight in metric — usually like "3 - 5"
    w = cat.get('weight', {}).get('metric')
    if w:
        parts = w.split(' - ')
        try:
            weights.append( (float(parts[0]) + float(parts[1]))/2 )
        except:
            pass

    ls = cat.get('life_span')
    if ls:
        parts = ls.split(' - ')
        try:
            lifespans.append( (float(parts[0]) + float(parts[1]))/2 )
        except:
            pass

weights = np.array(weights)
lifespans = np.array(lifespans)

def stats(arr):
    return {
        "min": float(np.min(arr)),
        "max": float(np.max(arr)),
        "mean": float(np.mean(arr)),
        "median": float(np.median(arr)),
        "std": float(np.std(arr))
    }

print("Cat weight (kg) stats:", stats(weights))
print("Cat lifespan (years) stats:", stats(lifespans))
```

```python
#  Countries API — 10 largest countries, top languages, total languages
import requests
from collections import Counter

# Correct API endpoint
countries_api = "https://restcountries.com/v3.1/all"

# Fetch
response = requests.get(countries_api)
countries = response.json()


# SAFE helper to get nested values
def get_name(country):
    try:
        return country["name"]["common"]
    except:
        return "Unknown"

def get_population(country):
    try:
        return country["population"]
    except:
        return 0

def get_languages(country):
    try:
        return list(country["languages"].values())
    except:
        return []


# TEN LARGEST COUNTRIES (POPULATION)
sorted_countries = sorted(
    countries,
    key=lambda c: get_population(c),
    reverse=True
)

top10 = sorted_countries[:10]

print("Top 10 Largest Countries:")
for c in top10:
    print(get_name(c), get_population(c))


#  TEN MOST SPOKEN LANGUAGES
lang_counter = Counter()

for c in countries:
    langs = get_languages(c)
    for lang in langs:
        lang_counter[lang] += 1

print("\nTop 10 Most Spoken Languages:")
print(lang_counter.most_common(10))


#  TOTAL NUMBER OF LANGUAGES
all_langs = set()

for c in countries:
    langs = get_languages(c)
    for lang in langs:
        all_langs.add(lang)

print("\nTotal number of languages:", len(all_langs))
```

```python
# performing the measure of central tendancy and dispersion
class Statistics:

    def __init__(self, data):
        self.data = data
        self.data_sorted = sorted(self.data)

    def count(self):
        return len(self.data)

    def sum(self):
        return sum(self.data)

    def min(self):
        return min(self.data)

    def max(self):
        return max(self.data)

    def range(self):
        return self.max() - self.min()

    def mean(self):
        return round(self.sum() / self.count())

    def median(self):
        n = self.count()
        mid = n // 2
        if n % 2 == 0:
            return (self.data_sorted[mid - 1] + self.data_sorted[mid]) / 2
        else:
            return self.data_sorted[mid]

    def mode(self):
        freq = {}
        for num in self.data:
            freq[num] = freq.get(num, 0) + 1

        mode_value = max(freq, key=freq.get)
        return (mode_value, freq[mode_value])

    def var(self):
        mean = self.mean()
        return round(sum((x - mean)**2 for x in self.data) / self.count(), 1)

    def std(self):
        return round(self.var() ** 0.5, 1)

    def freq_dist(self):
        freq = {}
        for num in self.data:
            freq[num] = freq.get(num, 0) + 1

        result = [(round((count / self.count()) * 100, 1), value)
                  for value, count in freq.items()]
        return sorted(result, reverse=True)

    def describe(self):
        return f"""Count: {self.count()}
Sum:  {self.sum()}
Min:  {self.min()}
Max:  {self.max()}
Range:  {self.range()}
Mean:  {self.mean()}
Median:  {self.median()}
Mode:  {self.mode()}
Variance:  {self.var()}
Standard Deviation:  {self.std()}
Frequency Distribution: {self.freq_dist()}"""


# TESTING
ages = [
    31, 26, 34, 37, 27, 26, 32, 32, 26, 27,
    27, 24, 32, 33, 27, 25, 26, 38, 37, 31,
    34, 24, 33, 29, 26
]

data = Statistics(ages)
print(data.describe())
```

```python
# performing personal accounts 
class PersonAccount:

    def __init__(self, firstname, lastname):
        self.firstname = firstname
        self.lastname = lastname
        self.incomes = {}
        self.expenses = {}

    def add_income(self, description, amount):
        self.incomes[description] = amount

    def add_expense(self, description, amount):
        self.expenses[description] = amount

    def total_income(self):
        return sum(self.incomes.values())

    def total_expense(self):
        return sum(self.expenses.values())

    def account_balance(self):
        return self.total_income() - self.total_expense()

    def account_info(self):
        return f"""
Account Holder: {self.firstname} {self.lastname}
Total Income: {self.total_income()}
Total Expense: {self.total_expense()}
Balance: {self.account_balance()}
"""


# TESTING
person = PersonAccount("Lucky", "Bata")

person.add_income("Salary", 50000)
person.add_income("Grant", 20000)

person.add_expense("Food", 8000)
person.add_expense("Transport", 3000)

print(person.account_info())
```

```python
#dataset tables scrapper
# uci_dataset_table_scraper.py

import requests
from bs4 import BeautifulSoup
import pandas as pd

# 1️⃣ URL of UCI datasets page
url = "https://archive.ics.uci.edu/ml/datasets.php"

# 2️⃣ Fetch the page
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# 3️⃣ Find all tables
tables = soup.find_all("table")

# 4️⃣ Look for tables that contain dataset info
dataset_rows = []

for table in tables:
    # Only process tables with more than 5 columns
    for row in table.find_all("tr"):
        cols = row.find_all("td")
        if len(cols) >= 5:
            # Extract dataset info
            dataset_name = cols[0].text.strip()
            data_type = cols[1].text.strip()
            task = cols[2].text.strip()
            attribute_type = cols[3].text.strip()
            instances = cols[4].text.strip()
            
            dataset_rows.append({
                "Dataset Name": dataset_name,
                "Data Type": data_type,
                "Task": task,
                "Attribute Type": attribute_type,
                "Instances": instances
            })

# 5️⃣ Convert to DataFrame
df = pd.DataFrame(dataset_rows)

# 6️⃣ Remove duplicates (just in case)
df.drop_duplicates(inplace=True)

# 7️⃣ Save to CSV
df.to_csv("uci_datasets_full.csv", index=False)

# 8️⃣ Show first 20 rows as a check
print("First 20 datasets from UCI ML repository:")
print(df.head(20))

print("\n✅ Data saved to 'uci_datasets_full.csv'")



```

```python
from flask import Flask, Response, request
import json
from bson.json_util import dumps
import pymongo
import os
import re # Regular expressions for cleaning text
from collections import Counter # For counting word frequency
from datetime import datetime

app = Flask(__name__)

# MongoDB Connection
MONGODB_URI='mongodb+srv://itstrevorndungi_db_user:Tahbeets1982@trevor.gbspcrh.mongodb.net/?appName=Trevor'
client = pymongo.MongoClient(MONGODB_URI)
db = client['Trevor']

@app.route('/')
def home():
    return "<h1>Student API is Running</h1><p>Visit <a href='/api/v1.0/students'>/api/v1.0/students</a> to see data.</p>"
@app.route('/api/v1.0/analyze', methods=['POST'])
def analyze_text():
    # 1. Get the text from the form data
    # We use .get() to avoid errors if 'text' is missing
    raw_text = request.form.get('text', '')

    if not raw_text:
        return Response(dumps({"error": "No text provided"}), status=400, mimetype='application/json')

    # 2. Analyze Characters
    char_count = len(raw_text)

    # 3. Clean the text (remove punctuation and make lowercase)
    # This regex removes everything except letters and spaces
    clean_text = re.sub(r'[^\w\s]', '', raw_text).lower()
    
    # 4. Analyze Words
    words = clean_text.split()
    word_count = len(words)

    # 5. Find Most Frequent Words
    # Counter creates a frequency map, .most_common(5) gives the top 5
    frequent_words = Counter(words).most_common(5)

    # 6. Prepare the result
    analysis_results = {
        "original_text": raw_text,
        "character_count": char_count,
        "word_count": word_count,
        "most_frequent_words": frequent_words,
        "analyzed_at": datetime.now()
    }

    # 7. Store in MongoDB (Analyses collection)
    db.analyses.insert_one(analysis_results)

    # 8. Return the response
    return Response(dumps(analysis_results), mimetype='application/json')

if __name__ == '__main__':
    port = int(os.environ.get("PORT", 5000))
    app.run(debug=True, host='0.0.0.0', port=port, use_reloader=False)
```
